```python
import pandas as pd
import numpy as np
```

```python
f = pd.read_csv("C://Users/aleks/Downloads/Telegram Desktop/use.csv")
f
```

```python
dic = dict()
dic["number_of_students"] = len(f)
dic["mean_math"] = np.mean(f["Математика"])
dic["mean_russian_for_physicists"] = np.mean(f[f["Физика"] == 100]["Русский язык"])
f["count"] = 1
dic["agg_math"] = f.groupby("Математика").agg({"Физика" : "max", "Русский язык" : "min"})
dic["agg_math_count"] = f.groupby("Математика").agg({"count" : "sum"})
f["coin"] = abs(f["Математика"] - f["Физика"]) + abs(f["Физика"] - f["Русский язык"]) + abs(f["Математика"] - f["Русский язык"])
dic["number_coincidence"] = len(f[f["coin"] == 0])
dic["sum"] = pd.Series(f["Математика"] + f["Физика"] + f["Русский язык"])
f["sumn"] = f["Математика"] + f["Физика"] + f["Русский язык"]
f["countsum"] = 1
s = f.groupby("sumn").agg({"countsum" : "sum", "sumn": lambda x : x})
dic["max_number_coincidence"] = max(s[s['countsum'] == 1]["sumn"])
dic
```
